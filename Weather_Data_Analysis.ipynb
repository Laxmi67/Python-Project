#**Calculate Statistics using NumPy**

>Convert Data to NumPy Arrays

```
temp = data['Temp_C'].values
dew_point_temp = data['Dew Point Temp_C'].values
rel_hum = data['Rel Hum_%'].values
wind_speed = data['Wind Speed_km/h'].values
visibility = data['Visibility_km'].values
pressure = data['Press_kPa'].values
```

> Calculate Mean

```
import numpy as np
mean_temp = np.mean(temp)
mean_dew_point_temp = np.mean(dew_point_temp)
mean_rel_hum = np.mean(rel_hum)
mean_wind_speed = np.mean(wind_speed)
mean_visibility = np.mean(visibility)
mean_pressure = np.mean(pressure)

print(f"Mean Temperature: {mean_temp:.2f}C")
print(f"Mean Dew Point Temperature: {mean_dew_point_temp:.2f}C")
print(f"Mean Relative Humidity: {mean_rel_hum:.2f}%")
print(f"Mean Wind Speed: {mean_wind_speed:.2f} km/h")
print(f"Mean Visibility: {mean_visibility:.2f} km")
print(f"Mean Pressure: {mean_pressure:.2f} kPa")
```

> Calculate Standard Deviation

```
std_temp = np.std(temp)
std_dew_point_temp = np.std(dew_point_temp)
std_rel_hum = np.std(rel_hum)
std_wind_speed = np.std(wind_speed)
std_visibility = np.std(visibility)
std_pressure = np.std(pressure)

print(f"Standard Deviation of Temperature: {std_temp:.2f}C")
print(f"Standard Deviation of Dew Point Temperature: {std_dew_point_temp:.2f}C")
print(f"Standard Deviation of Relative Humidity: {std_rel_hum:.2f}%")
print(f"Standard Deviation of Wind Speed: {std_wind_speed:.2f} km/h")
print(f"Standard Deviation of Visibility: {std_visibility:.2f} km")
print(f"Standard Deviation of Pressure: {std_pressure:.2f} kPa")
```

> Calculate Minimum and Maximum

```
min_temp = np.min(temp)
max_temp = np.max(temp)

min_dew_point_temp = np.min(dew_point_temp)
max_dew_point_temp = np.max(dew_point_temp)

min_rel_hum = np.min(rel_hum)
max_rel_hum = np.max(rel_hum)

min_wind_speed = np.min(wind_speed)
max_wind_speed = np.max(wind_speed)

min_visibility = np.min(visibility)
max_visibility = np.max(visibility)

min_pressure = np.min(pressure)
max_pressure = np.max(pressure)

print(f"Min Temperature: {min_temp:.2f}C, Max Temperature: {max_temp:.2f}C")
print(f"Min Dew Point Temperature: {min_dew_point_temp:.2f}C, Max Dew Point Temperature: {max_dew_point_temp:.2f}C")
print(f"Min Relative Humidity: {min_rel_hum:.2f}%, Max Relative Humidity: {max_rel_hum:.2f}%")
print(f"Min Wind Speed: {min_wind_speed:.2f} km/h, Max Wind Speed: {max_wind_speed:.2f} km/h")
print(f"Min Visibility: {min_visibility:.2f} km, Max Visibility: {max_visibility:.2f} km")
print(f"Min Pressure: {min_pressure:.2f} kPa, Max Pressure: {max_pressure:.2f} kPa")
```

># Temperature over the time.

> # Average temperature for each month

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('file.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%m/%d/%Y %H:%M')

df['Month'] = df['Date/Time'].dt.to_period('M')
df['Year'] = df['Date/Time'].dt.year

# Group by Month and calculate the average temperature
monthly_avg_temp = df.groupby('Month')['Temp_C'].mean()

plt.figure(figsize=(12, 6))
monthly_avg_temp.plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperature by Month')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],rotation=90)
plt.grid(axis='y')
plt.show()
```

> # Average relative humidity for each month

```
df = pd.read_csv('file.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df.set_index('Date/Time', inplace=True)
df['Month'] = df.index.month

# Group by month and calculate average humidity
monthly_avg_humidity = df.groupby('Month')['Rel Hum_%'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_humidity.plot(kind='bar', color='cyan')
plt.xlabel('Month')
plt.ylabel('Average Relative Humidity (%)')
plt.title('Average Relative Humidity by Month')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

> # Calculate the Frequency of Each Weather Condition

```
df = pd.read_csv('file.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)

# Calculate the frequency of each weather condition
weather_counts = df['Weather'].value_counts()

plt.figure(figsize=(15, 6))
weather_counts.plot(kind='bar', color='purple')
plt.xlabel('Weather Condition')
plt.ylabel('Frequency')
plt.title('Frequency of Different Weather Conditions')
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y')
plt.show()
```

 > # Wind speed changes over the Month

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('file.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)


df['Month'] = df.index.month
monthly_avg_wind_speed = df.groupby('Month')['Wind Speed_km/h'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_wind_speed.plot(kind='line', color='orange')
plt.xlabel('Month')
plt.ylabel('Average Wind Speed (km/h)')
plt.title('Average Wind Speed by Month')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y')
plt.show()
```
